# Test os.walk()


```python
import os

# Custom StructureError exception class
class StructureError(Exception):
    pass

directory_name = os.path.join(os.path.expanduser("~"), "src/personal/farzy.org/content")
```

```python
def walk_directory(path):
    for root, dirs, files in os.walk(path):
        depth = root.count(os.sep) - path.count(os.sep)
        print(f"Directory: {root}")
        print(f"Subdirectories: {dirs}")
        print(f"Files: {files}")
        print(f"Depth: {depth}\n")

walk_directory(directory_name)
```

# Use the os.walk() method to analyze a Hugo content directory

```python

def walk_hugo_directory(path):
    category_current = None
    for root, dirs, files in os.walk(path):
        depth = root.count(os.sep) - path.count(os.sep)
        print(f"Directory: {root}")
        if depth == 0:
            print(f"Categories: {', '.join(dirs)}\n")
        elif depth == 1:
            category_current = root.split(os.sep)[-1]
            print(f"Category: {category_current}")
            print(f"Articles: {', '.join(dirs)}\n")
        elif depth == 2:
            article_current = root.split(os.sep)[-1]
            print(f"Article: '{article_current}' in '{category_current}': {len(files)} languages.\n")
        elif depth == 3:
            print(f"Images for article '{article_current}': {', '.join(files)}\n")
        else:
             raise StructureError("Unexpected directory structure encountered.")
walk_hugo_directory(directory_name)
```
